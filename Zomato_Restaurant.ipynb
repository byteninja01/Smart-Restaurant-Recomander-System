```python
import pandas as pd
```

```python
delhi_ncr_restaurant_raw_data = pd.read_csv("/content/DelhiNCR Restaurants.csv",index_col=False)
```

```python
df = delhi_ncr_restaurant_raw_data.copy()
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Delivery_Rating'].fillna(df['Delivery_Rating'].mean(), inplace=True)
```

```python
df['Known_For2'].fillna('Unknown', inplace=True)
df['Known_For22'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(subset=['Restaurant_Name', 'Locality'], inplace=True)
```

```python
df['Phone_No'] = df['Phone_No'].astype(str).str.replace('[^0-9,]', '', regex=True)
```

```python
import numpy as np

q1 = df['Pricing_for_2'].quantile(0.05)
q3 = df['Pricing_for_2'].quantile(0.95)
df = df[(df['Pricing_for_2'] >= q1) & (df['Pricing_for_2'] <= q3)]
```

```python
df = df[(df['Dining_Rating'] >= 1.0) & (df['Dining_Rating'] <= 5.0)]
df = df[(df['Delivery_Rating'] >= 1.0) & (df['Delivery_Rating'] <= 5.0)]
```

```python
df['Average_Rating'] = df[['Dining_Rating', 'Delivery_Rating']].mean(axis=1)
```

```python
df['Total_Reviews'] = df['Dining_Review_Count'] + df['Delivery_Rating_Count']
```

```python
df['Known_For'] = df['Known_For2'] + ', ' + df['Known_For22']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Category_Encoded'] = le.fit_transform(df['Category'])
df['Locality_Encoded'] = le.fit_transform(df['Locality'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Pricing_for_2', 'Dining_Review_Count', 'Delivery_Rating_Count']] = scaler.fit_transform(
    df[['Pricing_for_2', 'Dining_Review_Count', 'Delivery_Rating_Count']]
)
```

```python
df.to_csv("zomato_processed.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("zomato_processed.csv")

plt.style.use('seaborn-v0_8')
sns.set_palette("viridis")
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nMissing values:\n", df.isnull().sum())
df.describe()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Dining_Rating'], bins=20, kde=True)
plt.title('Distribution of Dining Ratings')
plt.show()

plt.figure(figsize=(7,4))
sns.histplot(df['Delivery_Rating'], bins=20, kde=True, color='orange')
plt.title('Distribution of Delivery Ratings')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x='Pricing_for_2', y='Dining_Rating', data=df)
plt.title('Dining Rating vs Price for 2')
plt.show()

sns.scatterplot(x='Pricing_for_2', y='Delivery_Rating', data=df, color='red')
plt.title('Delivery Rating vs Price for 2')
plt.show()
```

```python
top_localities = df['Locality'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_localities.values, y=top_localities.index, palette='mako')
plt.title('Top 10 Localities by Restaurant Count')
plt.xlabel('Number of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_categories = df['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title('Most Common Cuisine Categories')
plt.xlabel('Number of Restaurants')
plt.show()
```

```python
locality_rating = df.groupby('Locality')['Dining_Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=locality_rating.values, y=locality_rating.index, palette='cool')
plt.title('Top 10 Localities by Average Dining Rating')
plt.xlabel('Average Rating')
plt.show()
```

```python
ratings = pd.melt(df[['Dining_Rating','Delivery_Rating']], var_name='Type', value_name='Rating')
plt.figure(figsize=(6,4))
sns.boxplot(x='Type', y='Rating', data=ratings)
plt.title('Comparison: Dining vs Delivery Ratings')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Dining_Rating','Delivery_Rating','Pricing_for_2','Dining_Review_Count','Delivery_Rating_Count','Total_Reviews']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig = px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    color="Dining_Rating",
    hover_name="Restaurant_Name",
    hover_data=["Locality", "Category", "Pricing_for_2"],
    color_continuous_scale="Viridis",
    size_max=10,
    zoom=11,
    height=500
)
fig.update_layout(mapbox_style="open-street-map", title="Restaurant Locations by Rating")
fig.show()
```

```python
df.groupby('Category')['Dining_Rating'].mean().sort_values(ascending=False).head(10).to_csv('top_categories.csv')
df.groupby('Locality')['Dining_Rating'].mean().sort_values(ascending=False).head(10).to_csv('top_localities.csv')
```
